```python
import pandas as pd

def dataexploration1(data):
    data.columns = ['CUSIP1', 'Name', 'Score']
    duplicates_data = data[data.duplicated()]
    data = data.drop_duplicates()
    na_data = data[data.isna()]
    data = data.dropna()
    print(data)
    data_mean = data['Score'].mean()
    print("Mean score is:", data_mean)
    data_std = data['Score'].std()
    print("Standard deviation of score is:", data_std)
    data_median = data['Score'].median()
    print("Median score is:", data_median)
    data_min = data['Score'].min()
    print("Minimum score is:", data_min)
    data_max = data['Score'].max()
    print("Maximum score is:", data_max)
    data_first_quantile = data.quantile(0.25)
    print("The location of the first quantile of the score column is:", data_first_quantile)
    data_third_quantile = data.quantile(0.75)
    print("The location of the third quantile of the score column is:", data_third_quantile)
    data_boxplot = data.boxplot(column='Score')
    print(data_boxplot)
    data_histogram = data.hist(column='Score', bins = 5)
    print(data_histogram)
    
data1 = pd.read_csv("MERTON_OUT20180731CY.CSV")
dataexploration1(data1)
```
